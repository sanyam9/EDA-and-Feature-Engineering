```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#Merging both the dataframes
final_df = pd.merge(df,df_country,on="Country Code",how='left')
final_df.head()
```

```python
# To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
#To find records with each country
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_count = final_df.Country.value_counts().values
```

```python
#Pie chart - Top 3 countries
plt.pie(country_count[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Observation: Maximum records are from India, then from the USA and then UK 

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

### Observations
1. When ratings are between 4.5 to 4.9 --> Excellent
2. When ratings are between 4.0 to 4.4 --> Very good
3. When ratings are between 3.5 to 3.9 --> Good
4. When ratings are between 2.5 to 3.4 --> Average
6. When ratings are between 1.0 to 2.4 --> Poor

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
/plt.show()
```

```python
#Plotting with colors given in the dataframe
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
plt.show()
```

Note: White is invisible

### Observations:
1. Not Rated count is very high
2. Maximum number of ratings are between 3.0 to 3.6

```python
#Count plot - useful for plotting categorical features
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```

Observation: Maximum number of 0 ratings are from Indian Customers
