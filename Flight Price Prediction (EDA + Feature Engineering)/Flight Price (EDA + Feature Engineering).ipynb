## Feature Engineering

```python
#Extracting Date,Month and Year from Date_of_Journey
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
#Fixing the data types
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']= df['Year'].astype(int)
```

```python
df.drop(columns='Date_of_Journey',inplace=True)
df.info()
```

```python
#Removing the date from Arrival_Time
df['Arrival_Time'] = df['Arrival_Time'].str.split(" ").str[0]
df.head()
```

```python
#Extracting hour and mins from Arrival time
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
#Fixing the data types
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
#Dropping Arrival Time
df.drop(columns='Arrival_Time',inplace=True)
df.head()
```

```python
#Extracting hour and mins from Departure time
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
#Fixing the data types
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
```

```python
#Dropping Dep Time
df.drop(columns='Dep_Time',inplace=True)
df.head()
```

```python
df.info()
```

Route and Total_stops conveys the same information. Therefore, we can drop the Route column and transform Total_stops into integer

```python
#Dropping route
df.drop(columns='Route',inplace=True)
```

```python
#Tranforming Total_stops
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].replace({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4},inplace=True)
```

```python
df.head()
```

```python
#Converting Duration into mins only 
df['Duration_hr']=df['Duration'].str.split(" ").str[0]
df['Duration_min']=df['Duration'].str.split(" ").str[1]
```

```python
#We have hours as 5 mins which needs to be dropped
df['Duration_hr'].unique()
```
